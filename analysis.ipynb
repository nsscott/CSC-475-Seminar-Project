```python
# data['How often do you use generative AI?']

options = ['daily', 'weekly', 'monthly', 'do not use']

# Create a python list of length 20 with random choices from options

import random
random_choices = [random.choice(options) for _ in range(20)]

random_choices

from matplotlib import pyplot as plt 

plt.hist(random_choices);
plt.xlabel('Frequency of Generative AI Use')
plt.ylabel('Number of Respondents')
plt.title('Survey on Generative AI Usage Frequency');
plt.savefig('generative_ai_usage_frequency.png')
```
