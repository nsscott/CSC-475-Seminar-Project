```python
# create a histogram of Frequency column

plt.figure(figsize=(10,6))
seminar_data['Frequency'].value_counts().plot(kind='barh')


# show percentage next to each bar with '%' sign
for index, value in enumerate(seminar_data['Frequency'].value_counts()):
    plt.text(value, index, f'{value} ({value/seminar_data.shape[0]*100:.1f}%)')


plt.title('Frequency of Generative AI Usage')
plt.ylabel('Frequency')
plt.xlabel('Number of Respondents')
plt.show()
```

```python
# Create a histogram of Use column 
plt.figure(figsize=(10,6))
seminar_data['Use'].value_counts().plot(kind='barh')
# show percentage next to each bar with '%' sign
for index, value in enumerate(seminar_data['Use'].value_counts()):
    plt.text(value, index, f'{value} ({value/seminar_data.shape[0]*100:.1f}%)')
plt.title('Broad Usage of Generative AI')
plt.ylabel('Usage')
plt.xlabel('Number of Respondents')
plt.show()
```

```python
# Create a histogram of AcademicUse column
plt.figure(figsize=(10,6))
seminar_data['AcademicUse'].value_counts().plot(kind='barh')

# Add newlines after every 100 characters on the y tick labels
# Split long ytick values into multiple lines each of say 50 characters so that they fit better
ytickvalues = seminar_data['AcademicUse'].value_counts().index
newtickvals = [val.split('()')[0].split(':')[0].split('.')[0] for val in ytickvalues]
plt.yticks(ticks=range(len(ytickvalues)), labels=newtickvals)

# show percentage next to each bar with '%' sign
for index, value in enumerate(seminar_data['AcademicUse'].value_counts()):
    plt.text(value, index, f'{value} ({value/seminar_data.shape[0]*100:.1f}%)')
plt.title('Typical Uses of Generative AI in Academic/Work Setting')
plt.ylabel('Academic/Work Usage')
plt.xlabel('Number of Respondents')
plt.show()
```
